```python
import pandas as pd
```

```python
df = pd.read_csv('../dataset/iris.csv')
```

```python
df
```

```python
df.columns
```

```python
x = df[['sepal.length', 'sepal.width', 'petal.length', 'petal.width']].values
```

```python
label_real = df['label'].values
```

----------------

```python
from sklearn.cluster import KMeans
```

```python
model_kmeans = KMeans(n_clusters=3, random_state=0).fit(x)
```

```python
label_cluster = model_kmeans.labels_
```

```python
labelcluster = []
for i in label_cluster:
    if i == 1:
        labelcluster.append(0)
    elif i == 2:
        labelcluster.append(1)
    elif i == 0:
        labelcluster.append(2)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
warna = sns.color_palette("viridis", 3, as_cmap=True)
```

```python
plt.figure(figsize=(10, 8))
plt.title('Dataset Asli')
sns.scatterplot(x=df['petal.length'], y=df['petal.width'], hue=label_real, palette=warna, s=60, linewidth=0, alpha=1)
plt.xlabel('Panjang Petal')
plt.ylabel('Lebar Petal')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
plt.title('Clustering dengan K-Means')
sns.scatterplot(x=df['petal.length'], y=df['petal.width'], hue=labelcluster, palette=warna, s=60, linewidth=0, alpha=1)
plt.xlabel('Panjang Petal')
plt.ylabel('Lebar Petal')
plt.show()
```
